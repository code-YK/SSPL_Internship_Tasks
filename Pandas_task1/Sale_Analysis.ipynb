```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'D:\internship\SSPL_Internship_Tasks\Pandas_task1\raw_sales_data.csv')
df
```

```python
df['Date']
```

```python
df['Date'] = pd.to_datetime(df['Date'],format='mixed',dayfirst=True)
```

```python
df
```

```python
df['Units_Sold']
```

```python
df['Units_Sold'] = df['Units_Sold'].apply(lambda x: np.nan if x < 0 else x)
```

```python
df['Revenue']
```

```python
df['Revenue'] = df['Revenue'].apply(lambda x: np.nan if x <= 0 else x)
```

```python
df['Holiday']
```

```python
df['Holiday'] = df['Holiday'].astype(str)
```

```python
df['Holiday'] = df['Holiday'].str.strip().str.upper() 
```

```python
yes_no_dict = { 'YES':'Yes' , 'Y':'Yes' , 'N':'No' , '-':'No' , 'NO':'No' }
df['Holiday'] = df['Holiday'].replace(yes_no_dict)
```

```python
df
```

```python
city_mode = df.groupby('Store_ID')['City'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else 'Unknown')
```

```python
df['City_mode'] = df['Store_ID'].map(city_mode)
```

```python
df['City'] = df['City'].fillna(df['City_mode'])
```

```python
df.drop('City_mode', axis=1, inplace=True)
```

```python
df
```

```python
df['Month'] = df['Date'].dt.month
```

```python
df.head()
```

```python
df['Is_Weekend'] = df['Date'].dt.dayofweek >= 5
```

```python
df.head()
```

```python
df['Avg_Selling_Price'] = df['Revenue'] / df['Units_Sold']
```

```python
df.head()
```

```python
monthly_revenue = df.groupby('Month')['Revenue'].sum().sort_index()
monthly_revenue
```

```python
top5_products = df.groupby('Product_ID')['Avg_Selling_Price'].mean().sort_values(ascending=False).head(5)
```

```python
top5_products
```

```python
correlation_matrix = df[['Units_Sold', 'Revenue', 'Avg_Selling_Price']].corr()
correlation_matrix
```
